```python
#hide
def draw_plot(graph_df,num_days,title):
    wb_df=graph_df.query("detected_state == '{0}'".format(title)).reset_index()
    wb_df=wb_df[["diagnosed_date",0]]
    df = wb_df.set_index(wb_df['diagnosed_date'])
    from datetime import date, timedelta
    dt = date.today() - timedelta(days=num_days)
    df=df.loc[dt:date.today()]
    p1=plt.bar(df['diagnosed_date'],df[0])
    plt.plot(df['diagnosed_date'],df[0])
    plt.ylabel('Number of Cases')
    plt.xlabel(title)
    plt.show()
    return
    
def get_states():
    states = [
    "Andhra Pradesh",
    "Arunachal Pradesh",
    "Assam",
    "Bihar",
    "Chhattisgarh",
    "Goa",
    "Gujarat",
    "Haryana",
    "Himachal Pradesh",
    "Jharkhand",
    "Karantaka",
    "Kerala",
    "Madhya Pradesh",
    "Maharashtra",
    "Manipur",
    "Meghalaya",
    "Mizoram",
    "Nagaland",
    "Odisha",
    "Punjab",
    "Rajasthan",
    "Sikkim",
    "Tamil Nadu",
    "Telangana",
    "Tripura",
    "Uttarakhand",
    "Uttar Pradesh",
    "West Bengal",
    "Andaman and Nicobar Islands",
    "Chandigarh",
    "Dadra and Nagar Haveli and Daman & Diu",
    "The Government of NCT of Delhi",
    "Jammu & Kashmir",
    "Ladakh",
    "Lakshadweep",
    "Puducherry"
    ]
    return states

import pandas as pd
#import numpy as np
import matplotlib.pyplot as plt
url = ('https://raw.githubusercontent.com/tashis/corona-in/master/IndividualDetails.csv')
states_df = pd.read_csv(url)
cols_keep = ["diagnosed_date", "detected_state", "current_status"]
states_red_df = states_df[cols_keep]
states_red_df['diagnosed_date']= pd.to_datetime(states_red_df["diagnosed_date"])
#states_red_df["diagnosed_date"] = pd.to_datetime(states_red_df["diagnosed_date"], format="%d/%m/%Y")
#df_cases_total = states_red_df.groupby(["detected_state","current_status"])[states_red_df['diagnosed_date']].size().to_frame().reset_index()
df_cases_total = states_red_df.groupby(["detected_state","current_status"]).size().to_frame().reset_index()
x=df_cases_total.query('current_status=="Hospitalized"')
x=x.reset_index()
x.columns
x=x[['detected_state', 0]]
y=df_cases_total.groupby('detected_state').sum()
y.rename(columns={0:'Cases'},inplace=True)
y=y.reset_index()
z=y.set_index('detected_state').join(x.set_index('detected_state'))
z.rename(columns={0:'Hospitalized'},inplace=True)
z=z.reset_index()
z.rename(columns={'detected_state':'State'},inplace=True)
graph_df=states_red_df.groupby(["detected_state","diagnosed_date"]).size().to_frame().reset_index()
```
