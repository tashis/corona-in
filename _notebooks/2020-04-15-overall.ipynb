```python
#hide
import pandas as pd

url = ('https://raw.githubusercontent.com/tashis/corona-in/master/covid_19_india.csv')
states_df = pd.read_csv(url)
cols_keep = ["Date", "State/UnionTerritory", "Confirmed", "Cured", "Deaths"]
states_red_df = states_df[cols_keep]
states_red_df["Date"]= pd.to_datetime(states_red_df['Date'])#, format="%d/%m/%Y")
df_cases_state = states_red_df[states_red_df.groupby('State/UnionTerritory').Date.transform('max') == states_red_df['Date']]
cols_keep = ["State/UnionTerritory", "Confirmed", "Cured", "Deaths"]
df_cases_state = df_cases_state[cols_keep]
df_cases_state = df_cases_state.reset_index()[cols_keep]
```

```python
#hide_input
import numpy as np
df_cases_state.loc['Total'] = df_cases_state.select_dtypes(np.number).sum()
df_cases_state
```
