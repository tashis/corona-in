```python
def draw_plot_alt(graph_df,num_days,title):
    import altair as alt
    wb_df=graph_df.query("detected_state == '{0}'".format(title)).reset_index()
    wb_df=wb_df[["diagnosed_date",0]]
    wb_df.rename(columns={0:'Cases'},inplace=True)
    wb_df=wb_df.groupby("diagnosed_date").sum().reset_index()
    #wb_df
    base = alt.Chart(wb_df).mark_bar(color='firebrick').encode(
        alt.X('diagnosed_date', title=title),
        alt.Y('sum(Cases)', title='Cases')
    )

    text = base.mark_text(
        align='left',
        baseline='middle',
        dx=3  # Nudges text to right so it doesn't appear on top of the bar
    ).encode(
        text='Cases:Q'
    )
    return base, text
```
